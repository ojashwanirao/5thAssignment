```python
#1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate a synthetic dataset
np.random.seed(0)
x = np.random.rand(100)
y = 2 * x + np.random.rand(100)

# Create a DataFrame
df = pd.DataFrame({'X Axis': x, 'Y Axis': y})

sns.scatterplot(data=df, x='X Axis', y='Y Axis')
plt.title('Scatter Plot of X and Y')
plt.show()
```

```python

#2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate a dataset of random numbers
data = np.random.normal(0, 1, 1000)

sns.histplot(data, kde=True)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Data')
plt.show()
```

```python

#3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.

import seaborn as sns
import matplotlib.pyplot as plt

# Create a dataset representing categories and their corresponding values
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 40, 30, 35, 20]

sns.barplot(x=categories, y=values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart of Categories and Values')
plt.show()
```

```python
#4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate a dataset with categories and numerical values
categories = ['A', 'B', 'C', 'D']
values = [np.random.normal(0, 1, 100) for _ in range(4)]

sns.boxplot(data=values, orient='h')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Box Plot of Values across Categories')
plt.show()
```

```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate a synthetic dataset with correlated features
np.random.seed(0)
data = np.random.rand(100, 5)
df = pd.DataFrame(data, columns=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4', 'Feature 5'])

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix of Features')
plt.show()
```
